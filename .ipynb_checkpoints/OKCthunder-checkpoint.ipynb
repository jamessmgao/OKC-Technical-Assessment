```python
import numpy as np
import pandas as pd

okc = pd.read_csv("shots_data.csv")
okc["x squared"] = okc["x"]**2
okc["y squared"] = okc["y"]**2

display(okc)
```

```python
teamstring = "Team A" #switch to "Team B" to get team B's statistics

print("================ " + teamstring + " ===================")
team = okc[okc["team"] == teamstring]

##2 POINT ZONE

print("2")
two = team[team["x squared"] + team["y squared"] <= 23.75**2]
two = two[abs(two["x"]) <= 22.0]
display(two)

##NC3 POINT ZONE

print("NC3")
nthree = team[team["x squared"] + team["y squared"] > 23.75**2]
nthree = nthree[nthree["y"] >= 7.8]
display(nthree)

##C3 Point ZONE

print("C3")
cthree = team[abs(team["x"]) > 22.0]
cthree = cthree[cthree["y"] < 7.8]
display(cthree)

#check if all shots total to original data total
print(len(two) + len(nthree) + len(cthree), len(team))


print("=================PERCENTAGES=================")
print(len(two)/len(team))
print(len(nthree)/len(team))
print(len(cthree)/len(team))

#check if zone percentages add up to 1.0
print(len(two)/len(team) + len(nthree)/len(team) + len(cthree)/len(team))

print("=================eFG=================")
print(len(two[two["fgmade"] == 1])/len(two))
print(len(nthree[nthree["fgmade"] == 1])*1.5/len(nthree)) #standardize
print(len(cthree[cthree["fgmade"] == 1])*1.5/len(cthree))
```
